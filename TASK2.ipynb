```python
#loading of the datasets
import pandas as pd
df=pd.read_csv("/content/Unemployment in India.csv")
```

```python
#displaying the all information
print(df.info())
print(df.describe())
#displaying the first five data
print(df.head())
#displaying the last five elements
print(df.tail())
print(df.columns)
```

```python
#cleaning of datasets
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
df.isnull().sum()
print(df.columns)
print(df[' Estimated Employed'].dtype)
print(df[' Estimated Unemployment Rate (%)'].dtype)
df[' Estimated Unemployment Rate (%)'] = df[' Estimated Unemployment Rate (%)'].fillna(df[' Estimated Unemployment Rate (%)'].mean())
df['Estimated Employed'] = df[' Estimated Employed'].fillna(df[' Estimated Employed'].mean())
df.drop_duplicates(inplace=True)
df[' Date']=pd.to_datetime(df[' Date'])
df[' Date']=pd.to_datetime(df[' Date'],format='%d-%m-%y')
df[' Date']=pd.to_datetime(df[' Date'],errors='coerce')
scaler=StandardScaler()
df[[' Estimated Employed', ' Estimated Unemployment Rate (%)',' Estimated Labour Participation Rate (%)']] = scaler.fit_transform(df[[' Estimated Employed', ' Estimated Unemployment Rate (%)',' Estimated Labour Participation Rate (%)']])
min_max_scaler=MinMaxScaler()
df[[' Estimated Employed', ' Estimated Unemployment Rate (%)',' Estimated Labour Participation Rate (%)']] = min_max_scaler.fit_transform(df[[' Estimated Employed', ' Estimated Unemployment Rate (%)',' Estimated Labour Participation Rate (%)']])
#DATA AFTER SCALING
print(df.head())
print(df.tail())
```

```python
#print(df.columns)
import pandas as pd
import matplotlib.pyplot as plt
#unemplyment rate over time
plt.figure(figsize=(14,7))
plt.subplot(2,2,1)
plt.plot(df[' Date'], df[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='b')
plt.title('Estimated Unemployment Rate Over Time')
plt.xlabel("unemployment rate")
plt.ylabel("date")
plt.title("unemployment rate over time")
plt.grid(True)
plt.xticks(rotation=90)

```
